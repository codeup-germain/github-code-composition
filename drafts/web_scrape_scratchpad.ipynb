```python
import pandas as pd
import re
import requests
from bs4 import BeautifulSoup
```

### Get a list of 100 trending repos 
#### The list comes from https://github.com/trending which is Github's list of top trending repos

#### First, will be the Daily Trending Repos for 10.28.21

```python
#this request is from https://github.com/trending/python?since=daily&spoken_language_code=en (which is the daily trending on 10/28/21 for english and python)

response = requests.get('https://github.com/trending/python?since=daily&spoken_language_code=en')
soup = BeautifulSoup(response.text)
repos = soup.select('.lh-condensed')
```

```python
repo = repos[0]
```

```python
repo.select_one('.href')
```

```python
def get_repo_list(repo):
   
    repo_name = repo.a.text.strip()
      
    return {'repo_name': repo_name}
```

```python
df_oct_28_21 = pd.DataFrame([get_repo_list(repo) for repo in repos])
```

```python
df_oct_28_21['repo_name'] = df_oct_28_21['repo_name'].str.replace(r'\n\n', '')
```

```python
df_oct_28_21['repo_name'] = df_oct_28_21['repo_name'].str.replace(r' ', '')
```

```python
list(df_oct_28_21.repo_name)
```

#### Next, will be the Weekly Trending Repos for 10.28.21

```python
#this request is from https://github.com/trending/python?since=weekly&spoken_language_code=en (which is the weekly trending on 10/28/21 for english and python)

response = requests.get('https://github.com/trending/python?since=weekly&spoken_language_code=en')
soup = BeautifulSoup(response.text)
repos = soup.select('.lh-condensed')
```

```python
df__weekly_oct_28_21 = pd.DataFrame([get_repo_list(repo) for repo in repos])
df__weekly_oct_28_21
```

#### Now, will be the Monthly Trending Repos for 10.28.21

```python
#this request is from https://github.com/trending/python?since=monthly&spoken_language_code=en (which is the monthly trending on 10/28/21 for english and python)

response = requests.get('https://github.com/trending/python?since=monthly&spoken_language_code=en')
soup = BeautifulSoup(response.text)
repos = soup.select('.lh-condensed')
```

```python
df__monthly_oct_28_21 = pd.DataFrame([get_repo_list(repo) for repo in repos])
df__monthly_oct_28_21
```

### Concat the df's together

```python
df_oct_28_all = pd.concat([df_oct_28_21, df__weekly_oct_28_21, df__monthly_oct_28_21], ignore_index=True)
df_oct_28_all
```

### Clean up the text

```python
df_oct_28_all['repo_name'] = df_oct_28_all['repo_name'].str.replace(r'\n\n', '')
```

```python
df_oct_28_all
```
